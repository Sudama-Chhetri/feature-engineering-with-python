# 5 Number Summary and Box Plot

##### minimum,maximum,median(Q2),Q1,Q3 and IQR

```python
import numpy as np

lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximun=np.quantile(lst_marks,[0,0.25,0.50,0.75,1])
print(minimum,Q1,median,Q3,maximun)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
print(lower_fence)
```

```python
higher_fence=Q3+1.5*(IQR)
print(higher_fence)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(lst_marks,color='teal')
plt.show()
```
